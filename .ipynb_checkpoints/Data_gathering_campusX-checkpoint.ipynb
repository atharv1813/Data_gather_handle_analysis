```python
import requests
import pandas as pd
import time

api_key = "681a3e983b7ce2296815c94ce8ef84f2"  
tmdb = pd.DataFrame() 

for i in range(1, 429):
    try:
        response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}', timeout=10)
        
        # Check if the response is successful
        if response.status_code == 200:
            # Extract the results and select specific columns
            temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity']]
            
            # Use pd.concat instead of append
            tmdb = pd.concat([tmdb, temp_df], ignore_index=True)
            time.sleep(1)  # Sleep for a second to avoid hitting rate limits
        else:
            print(f"Error on page {i}: {response.status_code}")
            break  # Stop if there is an error
    except requests.exceptions.RequestException as e:
        print(f"Request failed on page {i}: {e}")
        time.sleep(5)  # Wait before retrying

tmdb  # Final DataFrame containing all movie data
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
# webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text

import requests
from bs4 import BeautifulSoup
import time

url = 'https://www.ambitionbox.com/list-of-companies?page=1'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8'
}

for attempt in range(3):  # Try up to 3 times
    try:
        webpage = requests.get(url, headers=headers, timeout=10)
        webpage.raise_for_status()
        
        soup = BeautifulSoup(webpage.text, 'lxml')
        companies = soup.find_all('div', class_='company-content-wrapper')
        print(f"Number of companies found: {len(companies)}")
        
        for i, company in enumerate(companies, start=1):
            print(f"Company {i}: {company.get_text(strip=True)}")
        break  # Exit loop if successful

    except requests.exceptions.RequestException as e:
        print(f"Attempt {attempt + 1}: Error fetching the webpage: {e}")
        time.sleep(5)  # Wait before retrying
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
how_old=[]
no_of_employee=[]

for i in company:

  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  reviews.append(i.find('a' , class_='review-count').text.strip())
  ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
  how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
  no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())

df=pd.DataFrame({'name':name,
   'rating':rating,
   'reviews':reviews,
   'company_type':ctype,
   'Head_Quarters':hq,
   'Company_Age':how_old,
   'No_of_Employee':no_of_employee,
   })
```

```python
final=pd.DataFrame()
for j in range(1,1001):
  webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_='company-content-wrapper')
  name=[]
  rating=[]
  reviews=[]
  ctype=[]
  hq=[]
  how_old=[]
  no_of_employee=[]

  for i in company:

    try:
       name.append(i.find('h2').text.strip())
    except:
       name.append(np.nan)

    try:
       rating.append(i.find('p',class_='rating').text.strip())
    except:
       rating.append(np.nan)
   
    try:

      reviews.append(i.find('a' , class_='review-count').text.strip())
    except:
      reviews.append(np.nan)

    try:

      ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:

      hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    
    try:

      how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      how_old.append(np.nan)
    try:
      no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      no_of_employee.append(np.nan)
    

  df=pd.DataFrame({'name':name,
    'rating':rating,
    'reviews':reviews,
    'company_type':ctype,
    'Head_Quarters':hq,
    'Company_Age':how_old,
    'No_of_Employee':no_of_employee,
    })
  
  final = pd.concat([final,df],ignore_index=True)
```
